```python
import re
import pandas as pd

## setting up some sticky variables
parse_start = 'Dividends Only'
parse_end = 'djusted close price adjusted for both dividends and splits'
counter = 0
broken = 0
```

```python
# hacking first loop to create the dataframe
init_ticker = "AAT"
f = open(init_ticker + "_raw_file.txt", "r")
full_file = f.read()
f.close()

# trim out only the section with Dividend info
init_div_substring = ((full_file.split(parse_start))[1].split(parse_end)[0])
## stripping out the dates
init_div_dates = re.findall(r'\w\w\w\s\d\d\,\s\d\d\d\d',init_div_substring)
## stripping out the dividends
init_div_values = re.findall(r'\>\d+\.*\d*\<',init_div_substring)
init_div_values = [init_element[1:-1] for init_element in init_div_values]

master_div_frame = pd.DataFrame(
    {'ticker': init_ticker,
     'div_dates': init_div_dates,
     'div_values': init_div_values
    })
```

```python
f_list = open("ticker_list", "r")
for ticker in f_list:
    
    ticker = ticker.strip()
    
    # skipping due to the first dataframe hack above
    if ticker == "AAT":
        continue
    
    file_data_name = ticker + "_raw_file.txt"
    f_raw = open(file_data_name, "r")
    full_file = f_raw.read()
    f_raw.close()
    
    # trim out only the section with Dividend info
    div_substring = ((full_file.split(parse_start))[1].split(parse_end)[0])
    
    ## stripping out the dates
    div_dates = re.findall(r'\w\w\w\s\d\d\,\s\d\d\d\d',div_substring)
    
    ## stripping out the dividends
    div_values = re.findall(r'\>\d+\.*\d*\<',div_substring)
    div_values = [element[1:-1] for element in div_values]
    
    #validate lists are the same length
    if len(div_values) != len(div_dates):
        print("dropping ", ticker, " due to missing info")
        broken += 1
        continue
 
    print("ticker = ", ticker, "num dates = ", len(div_dates), "num values = ", len(div_values))
    counter += 1
    
    div_frame = pd.DataFrame(
        {'ticker': ticker,
         'div_dates': div_dates,
         'div_values': div_values
    })

    master_div_frame = master_div_frame.append(div_frame, ignore_index=True)

    print(div_frame)
       
```

```python
# carve out date ranges into 3 segment

## 10/14/2019 - 1/15/2020 --> NORM
## 2/21/2020 - 3/20/2020 --> DOWN
## 3/24/2020 - 4/10/2020 --> RECOVER
norm_sdate = date(2019, 10, 14)
norm_xdate1 = date(2019,12,24)
norm_xdate2 = date(2019,11,29)
norm_edate = date(2020, 1, 15)
down_sdate = date(2020, 2, 21)
down_edate = date(2020, 3, 20)
recov_sdate = date(2020, 3, 24)
recov_edate = date(2020, 4, 10)

norm_frame = full_frame[(pd.to_datetime(full_frame['date']) >= norm_sdate) 
                        & (pd.to_datetime(full_frame['date']) <= norm_edate)
                       & (pd.to_datetime(full_frame['date']) != norm_xdate1)
                       & (pd.to_datetime(full_frame['date']) != norm_xdate2)]
down_frame = full_frame[(pd.to_datetime(full_frame['date']) >= down_sdate) 
                        & (pd.to_datetime(full_frame['date']) <= down_edate)]
recov_frame = full_frame[(pd.to_datetime(full_frame['date']) >= recov_sdate) 
                         & (pd.to_datetime(full_frame['date']) <= recov_edate)]
```

```python
# drop master frame to free up memory space

full_frame = None
```

```python
# not elegant, however...
# getting index lists to calc correlations later
DJI_NORM_list = norm_frame[(norm_frame['ticker'] == "^DJI")]['delta']
#SNP_NORM_list = norm_frame[(norm_frame['ticker'] == "^GSPC")]['delta']
#NDQ_NORM_list = norm_frame[(norm_frame['ticker'] == "^IXIC")]['delta']
#TNX_NORM_list = norm_frame[(norm_frame['ticker'] == "^TNX")]['delta']
DJI_DOWN_list = down_frame[(down_frame['ticker'] == "^DJI")]['delta']
#SNP_DOWN_list = down_frame[(down_frame['ticker'] == "^GSPC")]['delta']
#NDQ_DOWN_list = down_frame[(down_frame['ticker'] == "^IXIC")]['delta']
#TNX_DOWN_list = down_frame[(down_frame['ticker'] == "^TNX")]['delta']
DJI_RECOV_list = recov_frame[(recov_frame['ticker'] == "^DJI")]['delta']
#SNP_RECOV_list = recov_frame[(recov_frame['ticker'] == "^GSPC")]['delta']
#NDQ_RECOV_list = recov_frame[(recov_frame['ticker'] == "^IXIC")]['delta']
#TNX_RECOV_list = recov_frame[(recov_frame['ticker'] == "^TNX")]['delta']
```

```python
# [norm_frame[(norm_frame['ticker'] == "AAPL")]['delta']]
# for loop in 

summary_rows = []
f = open("master_ticker_list", "r")
#f = open("ticker_list_short", "r")

for ticker in f:
    print("ticker = ", ticker)
    
    TICK_NORM_list = norm_frame[(norm_frame['ticker'] == ticker.strip())]['delta']
    TICK_DOWN_list = down_frame[(down_frame['ticker'] == ticker.strip())]['delta']
    TICK_RECOV_list = recov_frame[(recov_frame['ticker'] == ticker.strip())]['delta']
    
    if len(TICK_NORM_list) == len(DJI_NORM_list):
        DJI_NORM_corr_val = pearsonr(TICK_NORM_list,DJI_NORM_list)[0]
    if len(TICK_DOWN_list) == len(DJI_DOWN_list):
        DJI_DOWN_corr_val = pearsonr(TICK_DOWN_list,DJI_DOWN_list)[0]
    if len(TICK_RECOV_list) == len(DJI_RECOV_list):
        DJI_RECOV_corr_val = pearsonr(TICK_RECOV_list,DJI_RECOV_list)[0]

        summary_dict = {
            "ticker": ticker.strip(),
            "DJI_NORM": DJI_NORM_corr_val,
            "DJI_DOWN": DJI_DOWN_corr_val,
            "DJI_RECOV": DJI_RECOV_corr_val
        }
        summary_rows.append(summary_dict)
    
summ_frame = pd.DataFrame(summary_rows)
summ_frame.to_csv("summary_frame.csv")
```

```python
# irrational exhuberance, following in the recovery
summ_frame.loc[(summ_frame['DJI_NORM'] < -0.15) & (summ_frame['DJI_RECOV'] > 0.3)]
```

```python
import pycurl
import io

c = pycurl.Curl()
f = open("ticker_list", "r")
```

```python
for ticker in f:
    
    buffer = io.BytesIO()
    
    ticker = ticker.strip()
    print(ticker)
    local_url = "https://finance.yahoo.com/quote/"+ticker+"/history?period1=809308800&period2=1583712000&interval=div%7Csplit&filter=div&frequency=1d"
    print(local_url)

    c.setopt(c.URL, local_url)
    c.setopt(c.WRITEDATA, buffer)
    c.perform()
    c.getinfo(pycurl.HTTP_CODE)
    
    full_text = ""
    full_text = buffer.getvalue().decode()
    buffer.flush()
    buffer.close()
    
    file_out_local = ticker + "_raw_file.txt"
    f_out = open(file_out_local, "w")
    f_out.write(full_text)
    f_out.close()
```
